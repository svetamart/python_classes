### 1. Найти координаты точек пересечения

```python
from sympy import *

x = Symbol('x')

f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241


res = solve(f = g)
```

```python
x1 = res[0]
x1
```

```python
x2 = res[1]
x2
```

```python
x3 = res[2]
x3
```

```python
x4 = res[3]
x4
```

```python
y1 = f.subs(x, int(x1))
y1
```

```python
y2 = f.subs(x, int(x2))
y2
```

```python
y3 = f.subs(x, int(x3))
y3
```

```python
y4 = f.subs(x, int(x4))
y4
```

```python
pic1 = plot(f, g)
```
